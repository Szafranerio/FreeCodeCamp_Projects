```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/Users/bartlomiejszafran/Desktop/data/adult.data.csv')
```

```python
data.head()
```

```python
data.race.value_counts()
```

```python
avg_age_men = data[data['sex'] == 'Male'] ['age'].mean()
print(avg_age_men)
```

```python
num_bachelors = len(data[data['education'] == 'Bachelors'])
total_num = len(data)
percentage_bachelors = round(num_bachelors / total_num  * 100, 1)
print(percentage_bachelors)

```

```python
BMD_education = data[data['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
None_education = data[~data['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]
```

```python
higher = len(BMD_education[BMD_education.salary == '>50K'])
percentage_higher = round(higher / len(BMD_education) * 100, 1)

lower = len(None_education[None_education.salary == '>50K'])
percentage_lower = round(lower / len(None_education) * 100, 1)
print(percentage_higher)
print(percentage_lower)
```

```python
min_hours = data['hours-per-week'].min()
print(min_hours)
```

```python
filtered_data_over_50 = data[(data['salary'] == '>50K')]
print(filtered_data_over_50)
```

```python
filtered_data_over_50.head(25)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=filtered_data_over_50, x='native-country')
plt.xticks(rotation=90)
plt.xlabel('Native Country')
plt.ylabel('Count')
plt.title('Count of Individuals Over 50K by Native Country')
plt.tight_layout()
plt.show()
```

```python
filtered_data_over_50['native-country'].value_counts()
```

```python
filtered_data_over_50['native-country'].info()
```

```python
country_counts = filtered_data_over_50['native-country'].value_counts()
country_percentages = (country_counts / country_counts.sum()) * 100
print(country_percentages)
```

```python
india_high_income = filtered_data_over_50[(filtered_data_over_50['native-country'] == 'India') & (filtered_data_over_50['salary'] == '>50K')]
most_popular_occupation = india_high_income['occupation'].value_counts().idxmax()
print("The most popular occupation for individuals in India earning >50K is:", most_popular_occupation)
```
